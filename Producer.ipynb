```python
df = pd.read_csv("extracted_data_cleaned.csv")
df.head()
```

```python
df["Interview Country"] = df["Interview Country"].str.split("\n").str[0]
df = df.drop(columns = ['group', 'sender', 'date'])
```

```python
df.head()
```

```python
while True:
    dict_stock = df.sample(1).to_dict(orient="records")[0]
    producer.send('telebot-topic1', value=dict_stock)
    sleep(1)
```

```python
producer.flush() #clear data from kafka server
```
